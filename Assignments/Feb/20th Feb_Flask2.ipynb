```python
#Q1. Explain GET and POST methods.
'''
GET method sends data by appending the URL. Only limited data can be sent and is comparitively less
secure. It can be bookmarked and cached. It is efficient and so used more.

POST method sends data by appending the body. It can send large amounts of data and is comparitively more
secure. It cannot be bookmarked and cached. It is less efficient and so used less. 
'''
```

```python
#Q2. Why is request used in Flask?
'''
 Request is used in Flask to send data from the Client to Server.  This data can be
 recovered using the GET/POST Methods. Some important attributes of Request are form, args, cookies,
 files and method.
'''
```

```python
#Q3. Why is redirect() used in Flask?
'''
redirect() function is used in Flask to redirect the user to a specified URL with the status code.
It accepts location, status code and response as parameters.
'''
```

```python
#Q4. What are templates in Flask? Why is the render_template() function used?
'''
Templates are files that contain static data as well as placeholders for dynamic data. A template is 
rendered with specific data to produce a final document.Flask uses the Jinja template library to 
render templates.

render_template() function is used to generate output from a template file based on the Jinja2 engine
that is found in the application's templates folder. It renders the external HTML file instead of hardcoding
the HTML in the view function.
'''
```
