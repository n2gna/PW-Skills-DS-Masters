```python
#Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

'''The loc() function is label based data selecting method. This method includes the last element of the range passed in it, unlike iloc().
 loc() can accept the boolean data unlike iloc().'''

 '''The iloc() function is an indexed-based selecting method. This method does not include the last element of the range passed in it unlike loc().
 iloc() does not accept the boolean data unlike loc()'''
```

```python
'''Q3. Reindex the given dataframe using a variable, reindex = [3,0,1,2] and store it in the variable, new_df then find the output for both
 new_df.loc[2] and new_df.iloc[2]. Did you observe any difference in both the outputs? If so then explain it.'''


reindex = [3,0,1,2]
new_df = df.reindex(reindex)

print(new_df.loc[2],'\n')
print(new_df.iloc[2])

'''There is a difference in the output. loc() considers the input as label and returns the row with label 2. But iloc() considers the input as
index and returns the row that is at index 2 as per the default indexing convention (0,1,2,3.....).'''
```

```python
'''Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and calculates the difference between them in days,
 hours, and minutes using Pandas time delta. The program should prompt the user to enter the dates and display the result.'''

from datetime import datetime 

try:
    date1 = datetime.strptime(input('Enter the first date in the format YYYY-MM-DD'), '%Y-%m-%d')
    print(date1)
    date2 = datetime.strptime(input('Enter the second date in the format YYYY-MM-DD'), '%Y-%m-%d')
    print(date2)
    if date2 > date1:
        print(date2-date1)
    elif date1 > date2:
        print(date1-date2)
    else:
        print('0 days, 0:00:00')
except ValueError:
    print('Wrong date/format')

```

```python
'''Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified column to a categorical data type.
 The program should prompt the user to enter the file path, column name, and category order, and then display the sorted data'''


file_path = input('Enter the file path of the CSV file: ')
print(f'File: {file_path}')
col = input('Enter the column name')
print(f'Column: {col}')

df = pd.read_csv(file_path)
df[col] = df[col].astype('category')
print(f'Data type of {col} column: {df[col].dtypes}')
df.sort_values(col)

 
```
