```python
#Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

'''ZeroDivisionError, OverflowError and FloatingPointError are defined in the ArithmeticError class.'''

'ZeroDivisionError is raised when number is divided by zero.'
try:
    print(10/0)
except ZeroDivisionError as e:
    print(e)
    print('\n')
'OverflowError is raised when the stroing limit is exceeded'
try:
    n = 7.0
    for i in range(1,1000):
        n = n**i
        print(n)
except OverflowError as e:
    print(e)
```

```python
#Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

'LookupError is base class error of KeyError and IndexError.'

'Key Error is raised when a key is not found in a dictionary.'
try:
    d = {'name': 'Peter', 'age': 35}
    print(d['name'])
    print(d['job'])
except KeyError:
    print('KeyError')

'IndexError is raised when an index is not found in a sequence or list.'
try:
    l = ['apple', 'orange', 'mango', 'grape']
    print('\r')
    print(l[1])
    print(l[10])
except IndexError as e:
    print(e)
```
